## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

dado_1 = random.randint(1, 6)
dado_2 = random.randint(1, 6)
total_dados = dado_1 + dado_2
iterar = 1000


def probabilides (iterar): # como se cuando ponerle parametros si puedo dejar los parentesis vacios o cuales son los parametros
    contador = 0
    for total_dados in range (iterar): # creo un loop para que entre en "total_dados" 1000 veces
       if total_dados > 7 or total_dados % 2 == 0: #solo le pido que me traiga mayores de 7 o pares
              contador +=1 #le pido que vaya guarando los resultados
              porcentaje = contador / iterar # le pido que guarde en porcentaje lo que guarda en contador y lo divida por 1000
    return porcentaje

resultado = probabilides(iterar) #llamo a la funcion "probabildades" cuyo parametro es 1000 y lo guardo en "resultado"
print(resultado)
```

```python
import random

def calcular_probabilidad_simulada(tiradas=1000):
    
    sumas = [random.randint(1, 6) + random.randint(1, 6) for _ in range(tiradas)]
    
  
    casos_favorables = sum(1 for suma in sumas if suma > 7 or suma % 2 == 0)
    
    probabilidad = casos_favorables / tiradas
    return probabilidad

probabilidad = calcular_probabilidad_simulada()
print(f"La probabilidad de obtener un número mayor a 7 o par al lanzar dos dados (simulando 1000 veces) es: {probabilidad:.2f}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {} # esto es un lista? aqui se va a guardar bolas de cierto color em cierta posicion de la lista???

for i in range(60): #voy a iterar 60 veces
    if i < 10: #le pregunto a i cuanto vale y si me dice que vale 1 a 10 se trata de bolas blancas
        ball_box[i] = "White" # que quiere decir esa i entre corchetes a que se refiere??? 
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
```

```python

import numpy as np
ball_box = []

for i in range(60):
    if i < 10:  # Los primeros 10 elementos serán "White"
        ball_box.append("White")
    elif (i > 9) and (i < 30):  # Los siguientes 20 serán "Red"
        ball_box.append("Red")
    else:  # Los últimos 30 serán "Green"
        ball_box.append("Green")


n_simulaciones = 1000
            
def tomar_bola(n_simulaciones):
    contador_1 = 0 
    contador_2 = 0

    for i in range(n_simulaciones):
        colores = []

        for i in range(5):
            colores.append(ball_box[np.random.randint(0, 59)])

        colores = np.array(colores)
        
        bolas_blancas = sum(colores == "White")
        bolas_rojas = sum(colores == "Red")
        bolas_verdes = sum(colores == "Green")
       
        if (bolas_blancas == 3) and (bolas_rojas == 2):
            contador_1 += 1
        
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            contador_2 += 1
    
    return contador_1 / n_simulaciones, contador_2 / n_simulaciones

probabilities = tomar_bola(1000)

print(f"La probabilidad de que salgan 3 bolas blancas y 2 rojas es: {np.round(probabilities[0] * 100, 2)}%") # no comprendo la ultima parte del print ...
print(f"La probabilidad de que sean todas del mismo color es: {np.round(probabilities[1] * 100, 2)}%")
```

```python
simulaciones = 1000

def tirar_bola (simulaciones):
    contador_1 = 0
    contador_2 = 0

    for i in range (simulaciones):
        
        color = np.random.choice(ball_box, size=5, replace=True)

        bolas_blancas = np.count_nonzero(color == "White")
        bolas_rojas = np.count_nonzero(color == "Red")
        bolas_verdes = np.count_nonzero(color == "Green")

        if (bolas_blancas == 3) and (bolas_rojas == 2):
            contador_1 += 1
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            contador_2 += 1

    return contador_1 / simulaciones, contador_2 / simulaciones

posibilidades = tirar_bola(simulaciones)

print(posibilidades)
```
